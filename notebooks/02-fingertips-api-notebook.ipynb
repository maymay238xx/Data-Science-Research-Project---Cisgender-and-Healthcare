```python
admissions_candidates_df = phof_indicators_df[
    phof_indicators_df["Indicator"].str.contains(
        "admission|emergency|hospital|avoidable|preventable",
        case=False,
        na=False
    )
]

admissions_candidates_df[["Indicator ID", "Indicator"]].sort_values("Indicator").head(25)
```

This indicator helps explain healthcare cost differences by showing who uses emergency care more frequently.

```python
# Set chosen indicator
ADMISSIONS_INDICATOR_ID = 21001

# Pull data using API
admissions_raw_df = ftp.get_data_for_indicator_at_all_available_geographies(ADMISSIONS_INDICATOR_ID)

# Preview data
admissions_raw_df.head()
```

### Clean Data Set

```python
admissions_raw_df[['Time period', 'Time period Sortable', 'Time period range']].head(10)
```

```python
# Select analysis columns
admissions_clean_df = admissions_raw_df[
    ['Time period Sortable', 'Sex', 'Area Name', 'Value']
].copy()

# Rename columns
admissions_clean_df.columns = ['Year', 'Sex', 'Area', 'Value']

# Convert sortable year into numeric year
admissions_clean_df['Year'] = admissions_clean_df['Year'].astype(int) // 10000

# Ensure Value is numeric
admissions_clean_df['Value'] = pd.to_numeric(admissions_clean_df['Value'], errors='coerce')

# Filter to male vs female only
admissions_clean_df = admissions_clean_df[admissions_clean_df['Sex'].isin(['Male', 'Female'])]

# Preview clean data
admissions_clean_df.head()
```

Fingertips displays time as a formatted label (e.g. "2010/11") and a numeric sortable value. We used the sortable time field to safely extract numerical years for trend analysis.

```python
admissions_clean_df['Year'].unique()[:6]
```

```python
# Remove duplicate England entries
admissions_final_df = admissions_clean_df.drop_duplicates(
    subset=['Year', 'Sex', 'Area']
)

admissions_final_df.head()
```
